```python
import pandas as pd
import numpy as np
import glob


path      = r'/home/patrick/Documents/python_project/covid-nlp/lockdown/repull'
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)
print(li)
df = pd.concat(li, axis=0, ignore_index=True)
df.describe()
```

```python
df['removed_stopwords'].fillna(df['removed_noise'], inplace=True)
# df['sentiment'] = map(reduce df['removed_stopwords'])
# df.head()
from wordcloud import WordCloud

df.drop(df.index[8])
# Join the different processed titles together.
long_string = ','.join(list(df['removed_stopwords'].values))

# Create a WordCloud object
wordcloud = WordCloud(background_color="white", max_words=1000, contour_width=3, contour_color='steelblue')

# Generate a word cloud
wordcloud.generate(long_string)

# Visualize the word cloud
wordcloud.to_image()
```
